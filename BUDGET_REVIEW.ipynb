```python
EOY_Budget_csv = Path("Active_2021.csv")
EOY_Budget_df = pd.read_csv(EOY_Budget_csv)
EOY_Budget_df.head()
```

```python
EOY_Budget_df.dtypes
```

```python
EOY_Budget_df['VARIANCE_%']=EOY_Budget_df['VARIANCE']/EOY_Budget_df['PTD BUDGET']
EOY_Budget_df.head()
```

```python
budget_positive_gross = EOY_Budget_df.loc[(EOY_Budget_df['VARIANCE'] >= 10000) & (EOY_Budget_df['VARIANCE_%'] >= .10)] 
```

```python
budget_negative_gross = EOY_Budget_df.loc[(EOY_Budget_df['VARIANCE'] <= -10000) & (EOY_Budget_df['VARIANCE_%']) >= .10] 
```

```python
budget_join = [budget_positive_gross, budget_negative_gross]
```

```python
budget_variance_review= pd.concat(budget_join)
```

```python
budget_variance_review.head()
```

```python
Budget_Variance_Final = budget_variance_review[~budget_variance_review['GL CODE'].astype(str).str.startswith('4')]
```

```python
Budget_Variance_Final = Budget_Variance_Final[~Budget_Variance_Final['GL CODE'].astype(str).str.startswith('6')]
```
